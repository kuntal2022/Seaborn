# Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.


# ANS

Seaborn is a powerful data visualization library in Python that provides a high-level interface for creating informative and attractive statistical graphics. Here are five types of plots that can be created using Seaborn and their uses:

Line plot: A line plot displays data points connected by straight lines. It is used to show trends and changes over time.

Scatter plot: A scatter plot displays data points as individual dots on a two-dimensional plane. It is used to show the relationship between two variables.

Histogram: A histogram is a graph showing the distribution of numerical data. It is used to show the frequency distribution of a variable.

Heatmap: A heatmap is a graphical representation of data that uses color-coding to represent values. It is used to show the correlation between two or more variables.

Bar plot: A bar plot displays data as bars, where the height of each bar represents the value of the data. It is used to show comparisons between different groups or categories.

# Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
# "timepoint" and y = "signal" for different events and regions.
# Note: timepoint, signal, eve# nt, and region are columns in the fmri dataset.

```python
import seaborn as sns, pandas as pd, matplotlib.pyplot as plt

# Load the fmri dataset

fmri = sns.load_dataset("fmri")

# Create a line plot with event and region as hue
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", style="region")

# Show the plot
plt.show()
```

# Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.

```python
import seaborn as sns

# Load the titanic dataset
titanic = sns.load_dataset("titanic")

# Create a figure with two subplots
fig, axes = plt.subplots(ncols=2, figsize=(10, 6))

# Create a box plot of age by passenger class
sns.boxplot(data=titanic, x="pclass", y="age", ax=axes[0])
axes[0].set_title("Age by Passenger Class")

# Create a box plot of fare by passenger class
sns.boxplot(data=titanic, x="pclass", y="fare", ax=axes[1])
axes[1].set_title("Fare by Passenger Class")

# Show the plot
plt.show()
```

# Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
import seaborn as sns

# Load the diamonds dataset
diamonds = sns.load_dataset("diamonds")

# Plot a histogram of price with hue for cut
sns.histplot(data=diamonds, x="price", hue="cut")

# Show the plot
plt.show()
```

# Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.

```python
import seaborn as sns

# Load the iris dataset
iris = sns.load_dataset("iris")

# Plot a pair plot with hue for species
sns.pairplot(data=iris, hue="species")

# Show the plot
plt.show()
```
